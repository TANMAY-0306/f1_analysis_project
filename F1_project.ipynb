```python
def cleaning_qualifying_data (event_name):
    folder_name=f"{event_name.replace(' ','_')}"
    qualifying_data=pd.read_csv(f"C:/Users/TANMAY/code/f1_project_2025_datasets/{folder_name}/qualifying_raw_data.csv")
    q=qualifying_data
    q_shortlisted=q[['Driver','DriverNumber','LapTime','LapNumber','Stint','PitOutTime','PitInTime','Team','Position','Deleted','IsAccurate']] #shortlisting the working columns
    #cleaning_conditions------
    q1=q_shortlisted[q_shortlisted["IsAccurate"]==True]
    q2=q1[q1["Deleted"]==False]
    q3=q2[q2["LapTime"].notna()]
    q4=q3[q3['PitOutTime'].isna()]
    q5=q4[q4['PitInTime'].isna()]
    #cleaning done now computation of metric for evaluation------
    qualifying_metric=q5.groupby(["Driver","DriverNumber"]).LapTime.min() #this is a python series which is having driver as index not data to convert it into table we did the reset_index function which changed driver column into data----> a table
    qualifying_metric=qualifying_metric.reset_index()
    #now saving to a csv file-----
    qualifying_cleanfile_name="qualifying_cleaned_data.csv"
    qualifying_metric.to_csv(f"C:/Users/TANMAY/code/f1_project_2025_datasets/{folder_name}/{qualifying_cleanfile_name}",index=False)
    print(f"clean_qualifying_data for this {folder_name} has been created")
    
    
    
```

```python
def clean_race_data (event_name,event_format):
    #cleaning and computation for race results---
    folder_name=f"{event_name.replace(' ','_')}"
    race_results_data=pd.read_csv(f"C:/Users/TANMAY/code/f1_project_2025_datasets/{folder_name}/race_raw_results_data.csv")
    race_results_data=race_results_data[["DriverNumber","Abbreviation","TeamName","ClassifiedPosition","GridPosition","Status","Points","Laps"]]
    race_results_data.to_csv(f"C:/Users/TANMAY/code/f1_project_2025_datasets/{folder_name}/race_clean_results_data.csv")
    print(f"race result for{folder_name} has been created")

    #cleaning and computation for race laps----
    race_laps_data=pd.read_csv(f"C:/Users/TANMAY/code/f1_project_2025_datasets/{folder_name}/race_raw_laps_data.csv")
    r1=race_laps_data[race_laps_data["Deleted"]==False]
    r2=r1[r1["IsAccurate"]==True]
    r3=r2[r2["TrackStatus"]==1]
    r4=r3[r3["LapTime"].notna()]
    r5=r4[r4["PitOutTime"].isna()]
    r6=r5[r5["PitInTime"].isna()]
    # for computing 
    r6=r6.rename(columns={"LapTime":"RaceLapTime"})
    r6["LapTime"]=r6["RaceLapTime"]/pd.Timedelta(seconds=1)
    r7=r6.groupby(["Driver","DriverNumber"]).LapTime.median().reset_index()
    r8=r6.groupby(["Driver","DriverNumber"])
    r9=pd.DataFrame(columns=("Driver","DriverNumber","IQR"))
    for driver,race_data in r8:
        r_25=race_data.LapTime.quantile(0.25)
        r_75=race_data.LapTime.quantile(0.75)
        iqr=r_75-r_25
        Driver=driver[0]
        DriverNumber=driver[1]
        new_data=[Driver,DriverNumber,iqr]
        r9.loc[len(r9)]=new_data
    race_lap_metric=pd.merge(r9,r7,left_on=["Driver","DriverNumber"],right_on=["Driver","DriverNumber"])
    race_lap_metric.to_csv(f"C:/Users/TANMAY/code/f1_project_2025_datasets/{folder_name}/race_clean_laps_data.csv")
    print(f"race lap for{folder_name} has been created")
    
    
```

```python
def master_table_creation(event_name,event_format):
    #qualifying---->
    folder_name=f"{event_name.replace(' ','_')}"
    m_qualifying=pd.read_csv(f"C:/Users/TANMAY/code/f1_project_2025_datasets/{folder_name}/qualifying_cleaned_data.csv")
    m_qualifying=m_qualifying.rename(columns={"LapTime":"QualifyingLapTime"})
    m_qualifying["QualifyingLapTime"]=m_qualifying["QualifyingLapTime"]/pd.Timedelta(seconds=1)
    print(f"QUALIFYING TABLE DONE")
    # sprint------>
    if event_format=="sprint_qualifying":
        m_sprint=pd.read_csv(f"C:/Users/TANMAY/code/f1_project_2025_datasets/{folder_name}/sprint_cleaned_data.csv")
        m_sprint=m_sprint.rename(columns={"Abbreviation":"Driver","GridPosition":"SprintGridPosition","Status":"SprintStatus","Points":"SprintPoints"})
        m_sprint=m_sprint.drop(columns=["TeamName"])
        print(f"SPRINT TABLE FOR {folder_name} DONE")
    else:
        m_sprint=pd.DataFrame(columns=("DriverNumber","Driver","SprintPosition","SprintGridPosition","SprintStatus","SprintPoints"))
        print(f"SINCE THERE IS NO SPRINT IN {folder_name} EMPTY TABLE DONE")
        
        
    # race_laps----->
    m_race_laps=pd.read_csv(f"C:/Users/TANMAY/code/f1_project_2025_datasets/{folder_name}/race_clean_laps_data.csv")
    del m_race_laps["Unnamed: 0"]
    m_race_laps=m_race_laps.rename(columns={"LapTime":"RaceLapTime"})
    #race_results------>
    m_race_results=pd.read_csv(f"C:/Users/TANMAY/code/f1_project_2025_datasets/{folder_name}/race_clean_results_data.csv")
    del m_race_results["Unnamed: 0"]
    m_race_results=m_race_results.rename(columns={"Abbreviation":"Driver"})
    print(f"RACE TABLE DONE {folder_name}")

   
    #MASTER_TABLE_CREATION-------->
    m1=pd.merge(m_race_results, m_race_laps,left_on=["DriverNumber","Driver"],right_on=["DriverNumber","Driver"],how="left")
    m2=pd.merge(m1,m_sprint,left_on=["DriverNumber","Driver"],right_on=["DriverNumber","Driver"],how="left")
    master_table=pd.merge(m2,m_qualifying,left_on=["Driver","DriverNumber"],right_on=["Driver","DriverNumber"],how="left")
    
    #checks----->
    assert master_table.shape[0] == m_race_results.shape[0], \
    "Row count mismatch: master table and race results differ"


    assert not master_table.duplicated(
    subset=["Driver", "DriverNumber"]).any(), "Duplicate drivers found in master table"

    assert not master_table[["Driver", "DriverNumber"]].isna().any().any(), \
    "Missing Driver or DriverNumber after merge"

    if event_format == "sprint_qualifying":
        assert not master_table["SprintPosition"].isna().all(), \
        "Sprint weekend but sprint data missing"

    missing_quali = master_table["QualifyingLapTime"].isna().sum()
    print(f"Drivers without valid qualifying laps: {missing_quali}")
    
    master_table.to_csv(f"C:/Users/TANMAY/code/f1_project_2025_datasets/{folder_name}/master_table.csv",index=False)
    print(f"MASTER_TABLE FOR {folder_name} HAS BEEN CREATED ")
    
    
    

    

    
```

```python
for event in season_schedule.itertuples():
    event_name=event.EventName
    event_format=event.EventFormat
    master_table_creation(event_name,event_format)
    
```
