```python

def project_data_extract(round_number,event_name,event_format):
    season=2025
    folder_path=Path("C:/Users/TANMAY/code/f1_project_2025_datasets")
    folder_name=f"{event_name.replace(' ','_')}"
    path=folder_path/folder_name
    path.mkdir(exist_ok=True)
    print("folder_created")

    #now for extracting the qualifying raw data
    qualifying_session=f.get_session(season,event_name,"Q")
    qualifying_session.load()
    qualifying_raw_data=qualifying_session.laps
    name="qualifying_raw_data.csv"
    qualifying_raw_data.to_csv(f"C:/Users/TANMAY/code/f1_project_2025_datasets/{folder_name}/{name}",index=False)
    print("qualifying_data_extracted")
    
    #now for extracting the sprint raw data 
    if event_format=="sprint_qualifying":
        sprint_session=f.get_session(season,event_name,"S")
        sprint_session.load()
        sprint_raw_data=sprint_session.results
        name="sprint_raw_data.csv"
        sprint_raw_data.to_csv(f"C:/Users/TANMAY/code/f1_project_2025_datasets/{folder_name}/{name}",index=False)
        print("sprint_data_extracted")

    #now for extracting the race 
    race_session=f.get_session(season,event_name,"R")
    race_session.load()
    race_session_raw_results=race_session.results
    race_session_raw_laps=race_session.laps
    name1="race_raw_results_data.csv"
    name2="race_raw_laps_data.csv"
    race_session_raw_results.to_csv(f"C:/Users/TANMAY/code/f1_project_2025_datasets/{folder_name}/{name1}",index=False)
    race_session_raw_laps.to_csv(f"C:/Users/TANMAY/code/f1_project_2025_datasets/{folder_name}/{name2}",index=False)
    print("race_data_extracted")
```

```python
for event in season_schedule.itertuples():
    round_number=event.RoundNumber
    event_name=event.EventName
    event_format=event.EventFormat
    project_data_extract(round_number,event_name,event_format)
    #print(f"{round_number},{event_name},{event_format}")
      #22	United States	Las Vegas Grand Prix	conventional	Las Vegas
23	23	Qatar	Qatar Grand Prix	sprint_qualifying	Lusail
24	24	United Arab Emirates	Abu Dhabi Grand Prix	conventional	Yas Island

    
    
```
